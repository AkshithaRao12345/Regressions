Simple Linear Regression

Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Creating Dataset

```python
from random import choice

exp = []
sal = []

for i in range(15,60):
    exp.append(i + choice([1.6,-4,-2.3,4,1.9,-2.6,-1.7]))
    sal.append(36*i + 12.5 + choice([1.3,-4,-2.2,3.6,1.1,-2.5,-1.8]))      
    
df = pd.DataFrame(list(zip(exp,sal)), columns = "Experience Salary".split())
```

```python
df
```

```python
df.to_csv('expVsSal.csv', index = False)
plt.scatter(exp,sal)
```

Importing the dataset

```python
dataset = pd.read_csv("expVsSal.csv")

x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 0)
print(x_test)
print(y_test)
```

Training the Simple Linear Regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
obj = LinearRegression()
obj.fit(x_train, y_train)
```

Predicting the Test set results

```python
y_pred = obj.predict(x_test)
y_pred
```

```python
x_test
```

```python
y_test
```

Visualising the Training set results

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, obj.predict(x_train), color = 'red')
plt.title('Experience vs Salary (Training set)')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

Getting the final linear regression equation with the values of the coefficients

```python
print(obj.coef_)
print(obj.intercept_)
```
